# EXPLORATORY DATA ANALYSIS

**Gender**

```python
df.columns
```

```python
#using seaborn 
ax = sns.countplot(x= 'Gender', data =df) #countplot to count if X= Gender then the count will be displayed on Y-axis from data df
for bars in ax.containers: # to display the value in bars
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False) #group by gender bases , index false to not make the index in the original dataframe
#After grouping by gender, it calculates the total (sum) of the Amount column for each gender
# sort values: it sorts the resulting data by the Amount column in descending order, so the gender with the highest total sales appears first.
```

```python
sns.barplot(x='Gender', y='Amount' , data = sales_gen)
```

```python
# Now that we have the data, let's create the sales_state DataFrame to summarize top 10 states by Orders
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

# Creating a pie chart for the top 10 states by Orders
plt.figure(figsize=(8, 8))
plt.pie(sales_state['Orders'], labels=sales_state['State'], autopct='%1.1f%%', startangle=140)
plt.title('Top 10 States by Orders')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=10)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
age_group_orders = df.groupby('Age Group')['Orders'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(data=age_group_orders, x='Age Group', y='Orders', marker='o')
plt.title('Orders by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Orders')
plt.show()
```

```python
plt.figure(figsize=(16, 7))
sns.barplot(data=sales_state, x='State', y='Orders')
plt.title('Total Orders by State')
plt.xticks(rotation=45)
plt.show()
```

```python
age_group_orders = df.groupby('Age Group')['Orders'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(data=age_group_orders, x='Age Group', y='Orders', marker='o')
plt.title('Orders by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Orders')
plt.show()
```
